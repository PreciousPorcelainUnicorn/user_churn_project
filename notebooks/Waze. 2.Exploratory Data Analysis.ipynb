#### **Retention by kilometers driven per driving day**

```python
# 1. Create `km_per_driving_day` column
df['km_per_driving_day'] = df['driven_km_drives']/df['driving_days']

# 2. Call `describe()` on the new column
df['km_per_driving_day'].describe()
```

The mean value is infinity, the standard deviation is NaN, and the max value is infinity.
This is the result of there being values of zero in the `driving_days` column. Pandas imputes a value of infinity in the corresponding rows of the new column because division by zero is undefined.

```python
# 1. Convert infinite values to zero
df['km_per_driving_day'] = df['km_per_driving_day'].replace(np.inf, 0)

# 2. Confirm that it worked
df['km_per_driving_day'].describe()
```

The maximum value is 15,420 kilometers per drive day. This is physically impossible. Driving 100 km/hour for 12 hours is 1,200 km. It's unlikely many people averaged more than this each day they drove, so, for now, disregard rows where the distance in this column is greater than 1,200 km.

```python
df = df[df['km_per_driving_day'] <= 1200]
df.info()
```

```python
# Histogram

plt.figure(figsize=(10,6))
sns.histplot(data = df, x = 'km_per_driving_day', hue = 'label', multiple = 'fill')
plt.xlabel('Distance per driving day, km')
plt.ylabel('Number of users')
plt.show()
```

The churn rate tends to increase as the mean daily distance driven increases. It would be worth investigating further the reasons for long-distance users to discontinue using the app.

### Handling outliers

The box plots from the previous section indicated that many of these variables have outliers. It may be useful to impute outlying data with more reasonable values. One way of performing this imputation is to set a threshold based on a percentile of the distribution.

To practice this technique, write a function that calculates the 95th percentile of a given column, then imputes values > the 95th percentile with the value at the 95th percentile.

```python
def reassign_95(col):
    percentile95 = col.quantile(0.95)
    for val in col:
        if val > percentile95:
            col.replace(to_replace = val, value = percentile95, inplace = True)
    return col
```

```python
reassign_95(df['sessions'])
reassign_95(df['drives'])
reassign_95(df['total_sessions'])
reassign_95(df['driven_km_drives'])
reassign_95(df['duration_minutes_drives'])
```
