# Data exploration and hypothesis testing

**The purpose** of this project is to demostrate knowledge of how to conduct a two-sample hypothesis test.

**The goal** is to apply descriptive statistics and hypothesis testing in Python.

Research question:
Is there a statistically significant difference in mean amount of rides between iPhone® users and Android™ users?

### Imports and data loading




```python
# Import any relevant packages or libraries
import pandas as pd
import numpy as np
from scipy import stats
```

```python
# Load dataset into dataframe
df = pd.read_csv('waze_dataset.csv')
```

### Data exploration

```python
# 1. Create `map_dictionary`
map_dictionary = {'Android': 2, 'iPhone': 1}

# 2. Create new `device_type` column
df['device_type'] = df['device']

# 3. Map the new column to the dictionary
df['device_type'] = df['device_type'].map(map_dictionary)
df
```

```python
# Calculate the average number of drives for each device type
df.groupby(df['device'])['drives'].mean()
```


### Hypothesis testing

Hypotheses for this data project

H0: There is no difference in the average amount of drives between customers who use iPhones and Android devices.\
HA: There is a difference in the average amount of drives between customers who use iPhones and Android devices.

Choose 5% as the significance level and proceed with a two-sample t-test.

```python
# 1. Isolate the `drives` column for iPhone users.
iPhone_subset = df[df['device'] == 'iPhone']['drives']

# 2. Isolate the `drives` column for Android users.
Android_subset = df[df['device'] == 'Android']['drives']

# 3. Perform the t-test
stats.ttest_ind(a=iPhone_subset, b=Android_subset, equal_var=False)
```

The P-value is about 14.33%, above the significance level, so we fail to reject the null hypothesis. There is no statistically significant difference between iPhone and Android users' average number of drives.

## Conclusion

From the result of hypothesis test we can conclude that there is not statistically significant difference between iPhone and Android users' average number of drives. There is no need to improve user experience on a specific device.
